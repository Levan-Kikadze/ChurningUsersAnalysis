# User Churn and Campaign Analysis

This notebook analyzes user order data from the past year to understand patterns of user retention and the impact of promotional campaigns. Users are grouped by the recency of their last order (0–30, 31–60, 61–90, 91–120, and 120+ days since last order). For each group, the analysis explores:

- The share of users who used a discount or campaign on their last order versus those who did not.
- Which specific campaign types were used on users' last orders, including breakdowns for all users and for one-time (single-order) users.
- How campaign usage and organic (no campaign) behavior differ across recency cohorts.
- Visualizations and percentage breakdowns to highlight key trends in user reactivation and campaign effectiveness.

The goal is to identify which campaigns drive last orders, how discount dependency varies by user segment, and where opportunities exist to improve retention

# Campaign/No Campaign Breakdown and Analysis - All users

click on the left side of text to expand

I grouped users by the recency of their last order (0–30, 31–60, 61–90, 91–120, 120+ days) and looked at which campaign type they used on that final purchase. This tells us what actually “triggered” the last order before they went quiet.

# What stands out overall
AM Campaigns account for 31.6% of all last orders—our biggest lever by far. “No Campaign” is close behind at 28.1%, which is our organic baseline. ULC Activation (13.4%) and Delivery Fee discounts (13.2%) are the next meaningful blocks. Bolt Plus (5.3%) and ULC Engagement (3.9%) show up but are clearly secondary. Everything else is under 3% and likely either niche or poorly tagged.

Cohort reads

0–30 days (still warm): AM is 31.2%, No Campaign 28.3%. Bolt Plus is actually material here (12.2%), and ULC Engagement shows up at 7.1%. Activation is only 6.8%. Warm users either respond to broad AM promos or don’t need a promo at all. Plus and Engagement are visible only when people are still in the habit.

31–60 days (early lapse): AM and No Campaign are tied at 27.1%. Delivery Fee jumps to 18.9%. Activation climbs to 12.4%. Engagement dips to 5.5%. These users start to show price/fee sensitivity. We’re already pushing Activation, but not as aggressively as later cohorts.

61–90 days (mid lapse): AM peaks at 41.2%—we threw AM heavily at them. No Campaign is 25.8%. Activation is 13.8%. “Other Campaigns” spike to 4.9%. Bolt Plus basically disappears (0.4%). We used AM a lot, but these users still churned after that last order. Plus clearly isn’t the hook for this group.

91–120 days (deep lapse): Activation is 20.8% (its high point), Delivery Fee 20.6%, AM 26.3%, No Campaign 27.7%. Engagement is zero. Here we pivot hard to Activation and fee waivers. It gets an order, but not retention.

120+ days (very lapsed): AM 32.0%, No Campaign 29.9%, Activation 20.0%, DF 12.7%, “Other” 4.0%, Engagement 0.7%. The mix looks weirdly similar to the active cohort, which suggests we might be blanketing everyone with the same big promos over time. Engagement is basically absent here.

Interpretation and hypotheses

AM is a blunt but dominant tool. It drives many last orders, especially for 61–90 day lapsers, but clearly doesn’t fix retention by itself. It’s a “one more order” switch, not a loyalty strategy.

Our organic share (~26–30%) is stable. That’s the genuine demand baseline. Improvements in experience (speed, assortment, app UX) could nudge this up and lower promo burn.

Activation vs Engagement are misaligned. Activation is being used deep into the lapse window where we should be running true reactivation packages; Engagement barely shows up after 60 days, which means we’re not nurturing at-risk but still active users with the right mechanics.

Bolt Plus is front-loaded. It matters when people are fresh (0–30), then it vanishes. Either Plus members rarely reach “last order” status (good) or we’re not pushing Plus renewals/upsells later.

Delivery Fee discounts resonate with lapsers. They spike at 31–60 and 91–120 days. Waiving or capping fees may be a better hook for reactivation than % off food.

The “Other Campaigns” bucket becomes notable in 61–90 and 120+. We need to audit what lives there. If they’re random or one-offs, measure whether they did anything incremental or just added complexity.

What I’d do next

Run post-campaign survival analysis: after each campaign type, what % of users return in 30/60/90 days without another discount? Rank campaigns by incremental retention, not usage.

# Comparative Analysis: All Users vs. One-Time Users
(Focus: Campaign patterns in last orders)

1. What stands out overall
All Users (multi-order):

AM Campaigns dominate (31.6%), No Campaign is strong (28.1% organic).

Activation + Delivery Fee = 26.6% combined (core reactivation levers).

Bolt Plus (5.3%) and Engagement (3.9%) are secondary.

One-Time Users:

AM still leads but weaker (29.9% vs. 31.6%), No Campaign drops sharply (19.7% vs. 28.1%).

ULC Activation surges (26.6% → dominant #2) – heavy reliance for acquisition.

Delivery Fee discounts spike (15.1% vs. 13.2%) – fee waivers critical for first orders.

Bolt Plus collapses (1.0% vs. 5.3%) – irrelevant for one-time users.

Takeaway: One-time users are more campaign-dependent (80.3% used promos vs. 71.9% for all users). Activation + Delivery Fee = 41.7% of their behavior.

2. Cohort reads
0–30 days (warm)

All Users: Organic-heavy (28.3% No Campaign), diverse promos (Bolt Plus 12.2%, Engagement 7.1%).

One-Timers: Campaign-blitz (81.2% used promos):

ULC Activation = 26.7% (vs. 6.8% all users)

Bolt Plus = 3.8% (vs. 12.2%)

Zero Engagement usage.

31–60 days (early lapse)

All Users: AM = No Campaign (27.1% each), Delivery Fee surges (18.9%).

One-Timers: Activation dominates (31.2%) + Delivery Fee (17.7%). AM weak (25.3%).

61–90 days (mid lapse)

All Users: AM peaks (38.1%), "Other" promos emerge (4.9%).

One-Timers: AM spikes (38.0%) + "Other" promos double (11.8% vs. 4.9%). Activation still high (23.7%).

91–120 days (deep lapse)

All Users: Activation (20.8%) + Delivery Fee (20.6%) lead.

One-Timers: Delivery Fee soars (22.9%) + Activation (32.3%). No organic recovery (No Campaign = 15.2%).

120+ days (very lapsed)

All Users: Balanced AM (32.0%) + No Campaign (29.9%).

One-Timers: AM heavy (30.7%) + Activation (24.2%). Organic base weak (No Campaign = 22.6%).

3. Key contrasts & hypotheses
A. Acquisition vs. Retention tools:

ULC Activation: Used 2X more for one-timers (26.6% vs. 13.4%). Hypothesis: Activation acquires users but fails to retain them.

Bolt Plus: Only relevant for retained users (5.3% vs. 1.0%). Hypothesis: Effective loyalty driver but underused in first orders.

Engagement campaigns: Near-zero for one-timers (0.3% vs. 3.9%). Critical gap: Not engaging users early enough.

B. Discount sensitivity:

Delivery Fee discounts: 15.1% for one-timers vs. 13.2% all users. Strongest in 91–120d cohort (22.9%).

AM discounts: Less effective for one-timers (29.9% vs. 31.6%). Suggests food discounts alone don’t hook new users.

C. Organic baseline collapse:

No Campaign: 19.7% for one-timers vs. 28.1% all users. One-time users rarely order organically.

"Other" campaigns: Higher for one-timers (5.3% vs. 2.3%). Risk: Fragmented promos dilute effectiveness.

4. Interpretation & strategic implications
For one-time users:

We’re over-reliant on Activation (26.6%) and Delivery Fee waivers (15.1%) to acquire them.

Bolt Plus & Engagement are missing – likely why they don’t return.

Organic demand is weak (19.7% No Campaign) → experience improvements urgent.

For retained users:

AM + organic base sustains them (59.7% combined).

Bolt Plus/Engagement matter post-acquisition (9.2% combined).
